### Integer programming

```python
def decay(rate, t, starting_value):
    """
    Decay function
    :param rate:
    :param t:
    :return:
    """
    decayed_fmeasure = starting_value*(1.0 - rate)**t
    return decayed_fmeasure
```

```python
def get_time_given_decay(max_fmeasure, decayed_fmeasure, rate):
    """
    Retrieves time given decayed measure and decay rate (by inversion)
    :param decayed_measure:
    :param rate:
    :return:
    """
    t = math.log((decayed_fmeasure/max_fmeasure), 1.0-rate)

    return t
```

```python
def beta_rate(rate, rates):
    """
    Computes the normalized F-measure decay rate, which then is the Loss growth rate
    :param rate (float): decay rate
    :param rates (list): list of decay rates
    :return:
    """
    beta = rate / sum(rates)
    return beta
```

```python
def compute_loss(max_fmeasure, decayed_fmeasure, fsafe, rate, rates, est_duration):
    """
    Computes loss by estimating the decayed fmeasure given the decay rate after a set duration
    :param decayed_fmeasure:
    :param rate:
    :param duration:
    :return:
    """
    t0 = get_time_given_decay(max_fmeasure, decayed_fmeasure, rate)
    t = t0 + est_duration
    est_decayed_fmeasure = decay(rate, t, max_fmeasure)
    print("Current measure: {}. Estimated decayed measure: {}".format(decayed_fmeasure, est_decayed_fmeasure))

    if est_decayed_fmeasure < fsafe:
        diff = fsafe - est_decayed_fmeasure
    else:
        diff = 0

    beta = beta_rate(rate, rates)

    loss = beta*diff**2

    return float(loss)
```

```python
def compute_duration(distance, velocity, max_measure, curr_measure, restoration):
    """
    Computes duration of a mission to an area, which includes time to travel to the area given velocity, plus restoration 
    (whether to charge up battery or restore F-measure)
    """
    duration_travel = distance/velocity
    duration_restoration = restoration*(max_measure - curr_measure)
    duration = duration_travel + duration_restoration
    return duration
```

### Cost computation

```python
def convert_fmeasures_result2dict(pickled_fmeasures_result):
    """
    Converts the pickled result ADT into a 
    """
    areas = list(pickled_fmeasures_result.keys())
    length = len(pickled_fmeasures_result[areas[0]])
    
    converted_result = list() #Container of F-measure of areas for each timestep
    for i in range(length):
        d = dict()
        for area in areas:
            d[str(area)] = pickled_fmeasures_result[area][i]
        converted_result.append(d)
    return converted_result
```

```python
def compute_cost_fmeasures(fmeasures, fmax, fsafe, fcrit):
    """
    Computes the cost, (i.e., the sum of losses), given the F-measures
    :param fmeasures: dict of F-measures of areas
    :return:
    """
    computed_cost = 0
    areas = fmeasures.keys()
    for area in areas:
        if fmeasures[area] >= fsafe:
            loss = fmax - fmeasures[area]
        elif fcrit <= fmeasures[area] and fmeasures[area] < fsafe:
            loss = 2*(fmax - fmeasures[area])
        elif fmeasures[area] < fcrit:
            loss = (fmax - fmeasures[area])**2
        computed_cost += loss
    
    actual_cost = 0
    for area in areas:
        actual_diff = fmax - fmeasures[area]
        actual_cost += actual_diff
        
    return actual_cost, computed_cost
```

```python
def cost_history(converted_fmeasures_result, fmax, fsafe, fcrit):
    """
    Computes and compiles the cost throughout the history of operation
    Input:
        converted_fmeasures_result: (list) - list of fmeasures of areas for each t
    """
    computed_costs = list()
    actual_costs = list()
    total_computed_cost = 0
    total_actual_cost = 0
    for i in range(len(converted_fmeasures_result)):
        fmeasures = converted_fmeasures_result[i]
        cost = compute_cost_fmeasures(fmeasures, fmax, fsafe, fcrit)
        total_actual_cost += cost[0] #Actual cost
        actual_costs.append(cost[0])
        total_computed_cost += cost[1] #Computed cost
        computed_costs.append(cost[1])
#     print("Average cost: {}. SD: {}".format(np.mean(np.array(costs)), math.sqrt(np.var(np.array(costs)))))
#     print("Total cost:", total_cost)

    costs_dict = dict()
    costs_dict['actual'] = (total_actual_cost, actual_costs)
    costs_dict['computed'] = (total_computed_cost, computed_costs)
    return costs_dict
```
